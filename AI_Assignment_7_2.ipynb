**Prompt :** Fix the Loop Off-By-One error in the code. Give one or two example usage.

```python
def sum_upto_n(n):
  total = 0
  for  i in range(1,n+1):
    total += i
  return total


 # Example 1: Sum up to 5
result1 = sum_upto_n(5)
print(f"The sum up to 5 is: {result1}")

# Example 2: Sum up to 10
result2 = sum_upto_n(10)
print(f"The sum up to 10 is: {result2}")

```

**Explaination :**

**def sum_upto_n(n):** : This line defines a function named sum_upto_n that takes one argument, n.

**total = 0:** Inside the function, a variable total is initialized to 0. This variable will store the running sum.

**for i in range(1, n + 1):** : This is a for loop that iterates through a sequence of numbers. range(1, n + 1) generates numbers starting from 1 and going up to (and including) n. In each iteration, the current number is assigned to the variable i.

**total += i:** In each iteration of the loop, the current value of i is added to the total. This accumulates the sum of all numbers from 1 to n.

**return total:** After the loop finishes, the function returns the final value of total, which is the sum of all numbers from 1 to n.

**Prompt :** Fix the conditional logic Error in grading system and give one example.

```python
def grade_student(score):
  if score >= 70:
    return "A"
  elif score >= 40: # This covers scores from 40 to 69
    return "B"
  else: # This covers scores below 40
    return "C"

# Example usage:
score = 85
grade = grade_student(score)
print(f"A score of {score} gets a grade of: {grade}")
```

**Explaination :**

**def grade_student(score):**: This line defines a function named grade_student that takes one argument, score, which is expected to be a numerical value representing the student's score.

**if score >= 70:**: This is the first conditional check. If the score is greater than or equal to 70, the code inside this if block is executed.

**return "A":** If the score is 70 or above, the function immediately stops and returns the string "A".

**elif score >= 40:**: If the first condition (score >= 70) is false, the code moves to this elif (else if) condition. If the score is greater than or equal to 40 (and less than 70, because the first condition was false), the code inside this elif block is executed.

**return "B":** If the score is between 40 and 69 (inclusive), the function immediately stops and returns the string "B".

**else:**: If neither of the above conditions (score >= 70 or score >= 40) is true, the code in the else block is executed. This means the score is less than 40.

**return "C":** If the score is less than 40, the function immediately stops and returns the string "C".
