```python
# This function calculates the Fibonacci sequence up to a given number.
def calculate_fibonacci(limit):
  """Calculates the Fibonacci sequence up to a given limit."""
  # Initialize the first two Fibonacci numbers
  a = 0
  b = 1
  # Initialize the counter for the sequence
  count = 2
  # Initialize the list to store the Fibonacci sequence
  fibonacci_sequence = [a, b]

  # Generate the Fibonacci sequence
  while count <= limit:
    # Calculate the next Fibonacci number
    next_fib = a + b
    # Append the next Fibonacci number to the sequence
    fibonacci_sequence.append(next_fib)
    # Update the previous two Fibonacci numbers
    a = b
    b = next_fib
    # Increment the counter
    count += 1

  # Return the Fibonacci sequence
  return fibonacci_sequence

# Main function to demonstrate the Fibonacci calculation
def main():
  """Demonstrates the calculation of the Fibonacci sequence."""
  # Set the limit for the Fibonacci sequence
  number_limit = 10
  # Calculate the Fibonacci sequence
  fibonacci_result = calculate_fibonacci(number_limit)
  # Print the Fibonacci sequence
  print("Fibonacci series till", number_limit, ":", fibonacci_result)

# Execute the main function
main()
```
